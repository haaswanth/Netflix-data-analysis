# Data Cleaning

### Question 1: Which columns contain missing values in the dataset?



```python
df.isnull().sum()
```

### Observation

The dataset contains missing values in six columns:

- director: 2,634 missing values
- cast: 825 missing values
- country: 831 missing values
- date_added: 10 missing values
- rating: 4 missing values
- duration: 3 missing values

The remaining columns do not contain any missing values.

## Question 2: Remove the duplicate rows from the dataset?

```python
df.duplicated().sum()
```

### Observation:
There NO Duplicate rows in the Dataset

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.shape
```

## Question 3: Handle missing values in the dataset?

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['date_added'] = df['date_added'].fillna("Unknown")
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna("Unknown")
```

df.isnull().sum()

### Observation:
The dataset contained missing values in columns such as `director`, `cast`, `country`, `date_added`, `rating`, and `duration`. These missing values were handled by replacing them with appropriate values. After cleaning, all columns contain zero missing values, making the dataset ready for further analysis.

#### Final Observation

- The dataset was checked for missing values and duplicates.
- Missing values were handled appropriately.
- Duplicate rows were checked and no duplicate records were found.
- Data types and formatting were verified.
- After cleaning, the dataset is ready for further Exploratory Data Analysis (EDA).

# Data Analysis

### 1.What is the distribution of Movies and TV Shows on Netflix?

```python
df["type"].value_counts()
```

### Observation

- The analysis shows the distribution of content types available on Netflix.
- Movies make up the majority of the content, while TV Shows represent a smaller portion.
- This indicates that Netflix has a larger collection of movies compared to TV Shows in this dataset.

### 2.Which countries produce the most Netflix content?

```python
df["country"].value_counts().head(10)
```

### Observation

- The analysis shows the countries that contribute the highest number of Netflix titles.
- The United States produces the most Netflix content compared to other countries.
- Other countries also contribute a significant number of movies and TV shows, showing Netflix has a global content library.

### 3. Which ratings are most common on Netflix?

```python
df["rating"].value_counts().head(10)
```

### Observation

- The analysis shows the distribution of different content ratings available on Netflix.
- TV-MA is the most common rating, followed by TV-14 and TV-PG.
- This indicates that Netflix has a large amount of mature and teen-oriented content in its library.

### 4. Which year had the highest number of content releases?

```python
df["release_year"].value_counts().head(10)
```

### Observation

- The analysis shows the number of Netflix content releases across different years.
- The year 2018 had the highest number of content releases in the dataset.
- This indicates that Netflix significantly expanded its content library during this period.

### 5. How many Netflix titles are produced by each country?

```python
df.groupby("country")["title"].count().head(10)
```

### Observation

- The analysis shows the number of Netflix titles produced by different countries.
- The result displays the count of titles for each country in the dataset.
- It helps us understand the contribution of different countries to Netflix's content library.
